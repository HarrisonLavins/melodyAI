## Bach Chorale BWV 57-8
 Getting a feel for viewing/working with streams

```python
### Bach Chorale BWV 57-8 ###
# --------------------------------
# Getting a feel for viewing/working with streams
sBach = corpus.parse('bach/bwv57.8')
sBach.show('text')
```

```python
#Get the soprano part of the chorale
soprano = sBach.parts[0].measures(0,2)
soprano.show('text')
```

```python
#Iterate through soprano part and print all key changes
print('Key Changes:')
for keychange in soprano.recurse().getElementsByClass(key.Key):
    #Get key information and offset location
    print(keychange, keychange.offset)
    
print('-'*50)

#Iterate through soprano part and print all notes
print('Soprano Notes:')
for thisNote in soprano.recurse().getElementsByClass(note.Note):
    #Get note information and offset location
    print(thisNote)
```

```python
#Render the start of Bach score
sBach.measures(0,5).show()
```

```python
#Chordify and display
sChords = sBach.chordify()
sChords.measures(0,4).show()
```

```python
#Condense all chords into the middle range of treble clef 
for c in sChords.recurse().getElementsByClass('Chord'):
    c.closedPosition(forceOctave=4, inPlace=True)
    
sChords.measures(0,4).show()
```

```python
for c in sChords.recurse().getElementsByClass('Chord'):
    rn = roman.romanNumeralFromChord(c, key.Key('B-'))
    c.addLyric(str(rn.figure))
    
sChords.measures(0,4).show()
```

```python
#Collect all chords in a list
chordList = []

for c in sChords.flat:
    if 'Chord' not in c.classes:
        continue
    print(c.lyric, end=' ')
    chordList.append(c.lyric)
```

```python
print(chordList)
```

```python
#Store chord frequency counts in a dictionary
chords = dict.fromkeys(chordList, 0)

for c in chordList:
    chords[c] += 1
    
chords
```
